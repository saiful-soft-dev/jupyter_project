# 2G Cell Level KPI Calculation by Using Pandas eva() function!

```python
cell_level_kpi=raw.eval(
    '''
traffic=K3014+K3024
cssr=(1-CM30/K3003)*K3013A/(K3010A-A3129H)*100
cdr= (CM33/(K3013A+CH343+CH323-CH313-CH333))*100
hosr=(CH313+CH333)/(CH311+CH331)*100
tch=((K3011A+K3011B)/(K3010A+K3010B))*100
sd=(K3001/K3000)*100
    '''
    )
cell_level_kpi
```

# Create  a new DataFrame as new calculated KPI!

```python
new=pd.DataFrame()
new['Date']=cell_level_kpi.Date
new['Time']=cell_level_kpi.New_T
new['Cell']=cell_level_kpi.Cell
new['Traffic']=cell_level_kpi.traffic
new['CSSR']=cell_level_kpi.cssr
new['CDR']=cell_level_kpi.cdr
new['HOSR']=cell_level_kpi.hosr
new['TCH_Con']=cell_level_kpi.tch
new['SD_Con']=cell_level_kpi.sd
new
```

```python
new.to_csv('2G_Cell_Level_KPI_1.csv')
```
